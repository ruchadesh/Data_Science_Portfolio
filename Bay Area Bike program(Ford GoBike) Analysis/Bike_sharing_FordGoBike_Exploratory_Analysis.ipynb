### 2. Bike rides count across months

```python

bike_data_clean['start_time_month'] = bike_data_clean['start_time'].dt.month.astype(int)
```

```python
color = sb.color_palette()[0]
order = bike_data_clean['start_time_month'].value_counts().index
sb.countplot(data=bike_data_clean , x= 'start_time_month', color=color )
plt.xlabel('Month number')
plt.ylabel('Number of bike rides')
plt.title('Month Vs Bike ride count');
```

### Observations: 
- We can see a gradual increase in bike rides over the months from the starting of the year. It reached maximum value in October and dropped suddenly in Novembner and December
- This sudden drop could be because of the Holiday season. (people may go out of San Francisco during holidays / not go to office)

### 4. Bike ride usage by types of users 

```python
user_type_counts = bike_data_clean['user_type'].value_counts()
user_type_counts.shape[0]
```

```python
total_count = bike_data_clean.shape[0]
total_count
```

```python
#order = bike_data_clean['start_time_day_name'].value_counts().index
sb.countplot(data=bike_data_clean , x= 'user_type', color=color)
plt.xlabel('User type')
plt.ylabel('Number of bike rides')
plt.title('Bike rides usage across user types')
plt.ylim(0,1700000)
for i in range(user_type_counts.shape[0]):
    count = user_type_counts[i]
    pct_string = '{:0.2f}%'.format(100*count/total_count)
    plt.text(i , count+50000, pct_string , ha= 'center')
```

### Observation
- Out of total bike rides, ~89% of bike rides are done by subscribers and only ~11% of rides are done by customers. 
- The plot clearly shows that subscribers make use of this program a lot more than usual customers
